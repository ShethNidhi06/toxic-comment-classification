```python
politics_df = politics_df.apply(lambda x: x.str.lower() if x.dtype == "object" else x) 
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

politics_df['Tweet_punct'] = politics_df['tweet_text'].apply(lambda x: remove_punct(x))
politics_df.head(10)
```

```python

import numpy as np
import re
 
def clean_tweet(temp):
    if type(temp) == np.float64:
        return ""
    temp = temp.lower()
    temp=re.sub("b'","",temp)
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub("x[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
   
    return temp
```

```python
politics_df['clean_tweet'] = politics_df['Tweet_punct'].apply(lambda x:clean_tweet(x))
politics_df





```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

politics_df['Tweet_tokenized'] = politics_df['clean_tweet'].apply(lambda x: tokenization(x.lower()))
politics_df.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
politics_df['Tweet_nonstop'] = politics_df['Tweet_tokenized'].apply(lambda x: remove_stopwords(x))
politics_df.head(10)
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

politics_df['Tweet_lemmatized'] = politics_df['Tweet_nonstop'].apply(lambda x: lemmatizer(x))
politics_df.head()
```

```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

politics_df['Tweet_stemmed'] = politics_df['Tweet_lemmatized'].apply(lambda x: stemming(x))
politics_df.head()
```

```python
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
politics_df.to_csv('clean_dataset.csv',index=False)
 
```

```python
df=pd.read_csv("clean_dataset.csv")
```

```python
df['scores'] = df['clean_tweet'].apply(lambda review: sid.polarity_scores(review))

df.head()
```

```python
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])

df.head()
```

```python


df['comp_score'] = df['compound'].apply(lambda c: '1' if c >=0 else '0')



df.head()
```
